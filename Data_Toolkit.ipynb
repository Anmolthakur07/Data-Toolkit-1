6. What are the differences between NumPy arrays and Python lists?
 - Both NumPy arrays and Python lists can store multiple elements, but NumPy arrays are optimized for numerical computations.
 - 1. Performance & Speed:
  -  NumPy Arrays: Much faster because they use fixed data types and optimized C-based operations.
  - Python Lists: Slower because they store objects, requiring extra overhead.
 - 2. Memory Efficiency:
 - NumPy Arrays: Use less memory because they store elements of the same type in a contiguous block.
 - Python Lists: Use more memory due to dynamic type storage and additional pointers.
 - 3. Data Type Handling:
 - NumPy Arrays: Homogeneous (all elements must be of the same data type).
 - Python Lists: Heterogeneous (can contain multiple data types).
 -  4. Vectorized Operations vs. Loops:
 - NumPy Arrays: Allow vectorized operations (operations apply to the whole array at once).
 - Python Lists: Require explicit loops for element-wise operations.
 -  5. Built-in Mathematical Functions:
 - NumPy Arrays: Have built-in functions for statistics, algebra, and transformations.
 - Python Lists: Require manual implementations or loops.

```python
3. #Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np
# Generate random data points
np.random.seed(42)  # Ensures reproducibility
x = np.random.rand(50) * 10  # Random x values (0-10)
y = np.random.rand(50) * 100  # Random y values (0-100)
plt.scatter(x, y, color='blue', marker='o', alpha=0.7, edgecolors='black')

# Add labels and title
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Scatter Plot Example")

# Show plot
plt.show()

```

```python
4. #How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Sample dataset
np.random.seed(42)  # Ensures reproducibility
data = {
    "A": np.random.rand(10) * 100,
    "B": np.random.rand(10) * 50,
    "C": np.random.rand(10) * 75,
    "D": np.random.rand(10) * 20
}

# Create DataFrame
df = pd.DataFrame(data)
corr_matrix = df.corr()
print(corr_matrix)
# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add title
plt.title("Correlation Matrix Heatmap")

# Show plot
plt.show()

```

```python
6. #Create a DataFrame and add a new column based on an existing column.
import pandas as pd
# Sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Salary": [50000, 60000, 70000, 80000, 55000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame before adding the new column
print(df)
# Create a new column 'Tax' as 10% of 'Salary'
df["Tax"] = df["Salary"] * 0.10

# Display updated DataFrame
print(df)
df["Tax Category"] = df["Salary"].apply(lambda x: "High" if x > 65000 else "Low")
print(df)


```

```python
8. #Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np
# Create an array of x values
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10

# Define multiple y values (functions of x)
y1 = np.sin(x)  # Sine wave
y2 = np.cos(x)  # Cosine wave
y3 = np.tan(x)  # Tangent wave (with limits to avoid extreme values)
y3[y3 > 5] = np.nan  # Limit tangent values
y3[y3 < -5] = np.nan  # Limit tangent values
plt.figure(figsize=(8, 5))  # Set figure size

# Plot multiple lines
plt.plot(x, y1, label="Sine Wave", color="blue", linestyle="solid")
plt.plot(x, y2, label="Cosine Wave", color="red", linestyle="dashed")
plt.plot(x, y3, label="Tangent Wave", color="green", linestyle="dotted")

# Add labels, title, and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot Example")
plt.legend()  # Show legend

# Show grid and plot
plt.grid(True)
plt.show()
```

```python
10. #Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
# Generate random data following a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.randn(200) * 10 + 50  # Mean = 50, Std Dev = 10
plt.figure(figsize=(8, 5))  # Set figure size
sns.histplot(data, bins=20, kde=True, color="blue")

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Generated Data")

# Show plot
plt.show()
```

```python
11. # Perform matrix multiplication using NumPy.
import numpy as np
# Define two NumPy matrices (2x2)
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])
result = A @ B
print("Matrix Multiplication Result:\n", result)
result = np.dot(A, B)
print("Matrix Multiplication Result:\n", result)
result = np.matmul(A, B)
print("Matrix Multiplication Result:\n", result)
```

```python
13. #Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd
import numpy as np
# Generate random data
np.random.seed(42)  # For reproducibility
df = pd.DataFrame({
    "X": np.random.randn(100) * 10,  # Random X values
    "Y": np.random.randn(100) * 10,  # Random Y values
    "Z": np.random.randn(100) * 10,  # Random Z values
    "Category": np.random.choice(["A", "B", "C"], 100)  # Random categories
})
fig = px.scatter_3d(df, x="X", y="Y", z="Z", color="Category",
                     title="3D Scatter Plot Example",
                     size_max=10)

# Show the plot
fig.show()

```
